# Analyze sales data to identify trends

```python
# Assuming 'Date' is a column with date information
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)
sales_over_time = df['Sales'].resample('M').sum()
```

```python
# Assuming sales_over_time is a Series with dates as the index

# Plot sales over time
plt.figure(figsize=(10, 6))
plt.plot(sales_over_time)

# Add title and labels
plt.title('Sales Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

plt.show()
```

# Top-Selling Products

```python
top_products = df.groupby('Product')['Quantity Ordered'].sum().nlargest(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(top_products.index, top_products.values)

# Add labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black', fontweight='bold', fontsize=10)

# Add title and labels
plt.title('Top-Selling Products')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Display the chart
plt.show()
```

# Revenue Metrics for Business decision-making

```python
total_sales = df['Sales'].sum() / 1000  # Divide by 1000 to get the value in K
total_cost = df['Price Each'].sum() / 1000  # Divide by 1000 to get the value in K
total_profit = total_sales - total_cost
profit_margin = (total_profit / total_sales) * 100
print(f'Revenue Matrics are:')
print(f'Total Sales: ${total_sales:.2f}K')  # Display sales in K
print(f'Total Profit: ${total_profit:.2f}K')  # Display profit in K
print(f'Profit Margin: {profit_margin:.2f}%')
```

```python
# Create a pie chart for revenue breakdown by product category
revenue_by_category = df.groupby('Product')['Sales'].sum()
plt.figure(figsize=(10, 10))
plt.pie(revenue_by_category, labels=revenue_by_category.index, autopct='%1.1f%%')
plt.title('Revenue Breakdown by Category')
plt.axis('equal')
plt.show()
```
